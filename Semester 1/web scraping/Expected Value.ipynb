```python
import requests
from bs4 import BeautifulSoup

# Step 1: Send a request to the website and get the HTML content
url = 'https://collegescorecard.ed.gov/school/?128391-Western-Colorado-University&fos_code=1102&fos_credential=3'  # Replace with the URL of the website you're scraping

# Make a request to the website and check the status code
response = requests.get(url)
if response.status_code == 200:
    # Step 2: Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all the h2 elements with the specific class
    mydivs = soup.find_all("h2", class_="display-2 medium-light-blue-text font-weight-bold pb-3")

    # Loop through each element and print the text
    for blueText in mydivs:
        print(blueText.get_text(strip=True))  # get_text() extracts the inner text, and strip() removes any extra whitespace
else:
    print(f"Error: Unable to retrieve the page, status code {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Send a request to the website and get the HTML content
url = 'https://collegescorecard.ed.gov/school/?128391-Western-Colorado-University&fos_code=1102&fos_credential=3'  # Replace with the URL of the website you're scraping

# Make a request to the website and check the status code
response = requests.get(url)

if response.status_code == 200:
    # Step 2: Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Print the entire HTML to inspect the structure
    print(soup.prettify())  # This will print the entire HTML structure
    
    # Find all the h2 elements with the specific class
    mydivs = soup.find_all("div", class_="display-2 medium-light-blue-text font-weight-bold pb-3")
    
    # If nothing is found, print a message
    if not mydivs:
        print("No matching elements found with the given class name.")
    
    # Loop through each element and print the text
    for blueText in mydivs:
        print(blueText.get_text(strip=True))  # get_text() extracts the inner text, and strip() removes any extra whitespace
else:
    print(f"Error: Unable to retrieve the page, status code {response.status_code}")
```
