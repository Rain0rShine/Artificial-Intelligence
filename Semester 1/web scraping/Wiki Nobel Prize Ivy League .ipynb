```python
import requests
from bs4 import BeautifulSoup

# Function to check if the page contains "Princeton" or "Harvard"
def scrape_page(URL):
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, 'html.parser')
    str_soup = str(soup)
    
    if "Princeton" in str_soup or "Harvard" in str_soup:
        print(f"Ivy League University found on {URL}")
        return True
    return False

# Step 1: Send a request to the website and get the HTML content
url = 'https://en.wikipedia.org/wiki/List_of_Nobel_laureates'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Step 2: Find the tables that list Nobel laureates
tables = soup.find_all('table', {'class': 'wikitable'})

# Counters for Ivy and Non-Ivy laureates
countIvy = 0
countNotIvy = 0

# Step 3: Loop through each table and extract laureates' names and links
for table in tables:
    rows = table.find_all('tr')

    for row in rows:
        cols = row.find_all('td')
        
        if len(cols) > 1:
            # Find the link (<a>) in the second column (usually the name of the laureate)
            name_tag = cols[1].find('a')
            if name_tag:
                name = name_tag.get_text()
                link = 'https://en.wikipedia.org' + name_tag.get('href')

                # Step 4: Visit each laureate's page and check for "Princeton" or "Harvard"
                if scrape_page(link):
                    print(f"Ivy League found for {name}: {link}")
                    countIvy += 1  # Increment the Ivy count
                else:
                    print(f"Ivy League not found for {name}.")
                    countNotIvy += 1  # Increment the non-Ivy count

# Step 5: Print the proportion of Ivy League laureates
if (countIvy + countNotIvy) > 0:  # Prevent division by zero
    proportion = countIvy / (countIvy + countNotIvy)
    print(f"Proportion of Ivy League Nobel Laureates: {proportion:.2f}")
else:
    print("No laureates found.")
```
